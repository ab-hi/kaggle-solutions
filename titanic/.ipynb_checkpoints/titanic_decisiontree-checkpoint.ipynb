```python
import pandas as pd
import numpy as np

from sklearn import tree
from sklearn import ensemble
```

```python
data= pd.read_csv('./titanic.csv')
data.head(n=10)
```

```python
data.columns
```

```python
mean_age= int(data['Age'].mean())
data['Age']= data['Age'].fillna(-0.5)
```

```python
sex_mapping = {
    'male':0,
    'female':1
}

sex=data['Sex'].values

quant_sex = np.zeros(sex.shape[0])

for ix in range(sex.shape[0]):
    quant_sex[ix]=sex_mapping[sex[ix]]
```

```python
ds = np.asarray([data['Pclass'], quant_sex, data['Age'], data['SibSp'], data['Parch'],data['Fare'],data['Survived']]).astype('float').T

y = data['Survived'].values
cols = ['pclass', 'sex', 'Age', 'Sibsp', 'parch', 'fare', 'survived']
print cols
ds[2]
```

```python
split = int(ds.shape[0]*0.8)

x_train = ds[:split, :-1]
x_test = ds[split:, :-1]

y_train = ds[:split, -1]
y_test = ds[split: , -1]

print x_test.shape, x_train.shape
print y_test.shape, y_train.shape
```

```python
clf= tree.DecisionTreeClassifier()
```

```python
hist = clf.fit(x_train, y_train)
```

```python
i=0.00
for ix in range(x_test.shape[0]):
    if (clf.predict(x_test[ix]) == y_test[ix]):
        i = i+1
acc = i/x_test.shape[0]
print acc
```

```python
clf1 = ensemble.RandomForestClassifier(verbose =2 ,n_estimators=20) 
```

```python
clf1.fit(x_train,y_train)
```

```python
i1=0.00
for ix in range(x_train.shape[0]):
    if (clf1.predict(x_train[ix]) == y_train[ix]):
        i1 = i1+1
train_acc = i1/x_train.shape[0]



i2=0.00
for ix in range(x_test.shape[0]):
    if (clf1.predict(x_test[ix]) == y_test[ix]):
        i2 = i2+1
test_acc = i2/x_test.shape[0]
```
